### Test *circle*

```python
circle = pd.DataFrame(np.array([[-1, 0, 1], [0, 1, 2], [1, 0, 3], [0, -1, 4]]), columns = ['x', 'y', 'color'])
point = pd.DataFrame(np.array([[0, 0]]), columns=['x', 'y'])
```

```python
sns.scatterplot(x = circle.x, y = circle.y, hue = circle.color, palette=sns.color_palette("husl", 4), legend = None)
plt.scatter(x = 0, y = 0, color = 'black', marker = 'P')
```

*Przypadek gdy mamy 4 punkty w takiej samej odległości od punktu testowego (czarny krzyżyk) i każdy punkt jest innej klasy. Dla k = 1 powinniśmy otrzymać wektor 1 (lub różne - w zależności od implementacji), dla k = 2 wektor zawierający 1 i 2 (lub różne), dla k = 3 wektor zawierający 1, 2 i 3 (lub różne), a dla k = 4 wektor 1, 2, 3, 4. Otrzymane wektory powinny być losowe, czyli za każdym razem możemy otrzymać inny wynik.*

```python
# Jeżeli Twoja metoda wymaga innego formatu danych (lub inaczej nazywają się argumenty funkcji lub przyjmuje je w innej kolejności) 
# odpowiednio uzupełnij funkcję knn().

for k in [1, 2, 3, 4]:
    circle_pred = []
    for i in range(10):
        circle_pred.append(knn(X = circle.drop(["color"], axis = 1),  y = circle.color, Z = point, k = k))
    print("k: ", str(k), circle_pred)
```
